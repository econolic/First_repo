```python
import pandas as pd

# 1. Прочитання файлу за допомогою read_csv
df = pd.read_csv('2017_jun_final.csv')
```

```python
# 2. Виведення перших 5 рядків таблиці
print("Перші 5 рядків таблиці:")
display(df.head())
```

```python
# 3. Визначення розміру таблиці (кількість рядків та стовпців)
print("Розмір таблиці:", df.shape)
```

```python
# 4. Визначення типів всіх стовпців
print("Типи стовпців:")
print(df.dtypes)
```

```python
# 5. Розрахунок частки пропущених значень у кожній колонці
missing_fraction = df.isnull().sum() / len(df)
print("Частка пропущених значень у кожній колонці:")
print(missing_fraction)
```

```python
# 6. Видалення усіх стовпців, які містять пропуски, за виключенням "Мова.програмування"
cols_to_drop = [col for col in df.columns if (df[col].isnull().sum() > 0) and (col != 'Язык.программирования')]
df_clean = df.drop(columns=cols_to_drop)
print("Стовпці після видалення (залишається 'Язык.программирования' та ті, що не містять пропусків):")
print(df_clean.columns)

# 7. Перерахунок частки пропусків у кожній колонці після видалення
missing_fraction_clean = df_clean.isnull().sum() / len(df_clean)
print("Частка пропущених значень після видалення стовпців:")
print(missing_fraction_clean)
```

```python
# 8. Видалення усіх рядків, де є пропущені значення, з використанням dropna
df_dropna = df.dropna()
print("Розмір таблиці після видалення рядків з пропусками:", df_dropna.shape)

# 9. Визначення нового розміру таблиці
print("Новий розмір таблиці:", df_dropna.shape)

# 10. Створення таблиці python_data з рядками, де 'Мова.програмування' дорівнює 'Python'
python_data = df[df['Язык.программирования'] == 'Python']
print("Розмір таблиці з даними для Python:", python_data.shape)

# 11. Визначення розміру таблиці python_data
print("Розмір таблиці python_data:", python_data.shape)
```

```python
# 12. Групування даних за стовпцем "Посада"
grouped = df.groupby('Должность')

# 13. Агрегація мінімального та максимального значення у стовпчику "Зарплата.в.місяць"
# Переконаємося, що значення зарплати є числовими (якщо необхідно, перетворюємо)
df['Зарплата.в.месяц'] = pd.to_numeric(df['Зарплата.в.месяц'], errors='coerce')
agg_salary = df.groupby('Должность')['Зарплата.в.месяц'].agg(['min', 'max'])
print("Агреговані значення зарплати (мінімальне та максимальне) за посадою:")
display(agg_salary.sort_values(by='max', ascending=True))
```

```python
# 14. Створення функції, що обчислює середнє значення зарплати
def fill_avg_salary(row):
    return (row['min'] + row['max']) / 2

# Застосування функції до агрегованої таблиці та створення нового стовпця "avg"
agg_salary['avg'] = agg_salary.apply(fill_avg_salary, axis=1)
print("Агреговані дані із доданим стовпцем 'avg':")
display(agg_salary.sort_values(by='avg', ascending=True))

# 15. Описова статистика для нового стовпця "avg"
stats_avg = agg_salary['avg'].describe()
print("Описова статистика для стовпця 'avg':")
display(stats_avg)
```

```python
# 16. Збереження отриманої таблиці у CSV файл
agg_salary.to_csv('aggregated_salary_by_position.csv', index=True)
print("Результати збережено у файл 'aggregated_salary_by_position.csv'")
```
